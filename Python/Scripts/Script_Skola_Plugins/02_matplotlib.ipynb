### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
#Object-oriented approach is generally recommended for most use cases because:
#It provides more flexibility, especially when dealing with multiple subplots or complex plots.
#It allows better control over figure and axis management, avoiding potential issues with the stateful nature of the pyplot interface (e.g., when handling multiple plots in a single script or interactive environment).
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure: The overall window or canvas containing the plot.
# Axes: The area within the figure where the data is plotted (can contain multiple plots).
# Axis: The x or y scale of the plot within an axes object.
# Artist: Any element that is drawn on the axes (lines, titles, labels, etc.).
```

    ### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Data
x = np.linspace(-4, 4, 100)
y = x**2

# Object-oriented approach
fig, ax = plt.subplots()  # Create figure and axes
ax.plot(x, y)  # Plot the function

# Add title and axis labels
ax.set_title('Plot of y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Show the plot
plt.show()
```

```python
# Data
x = np.linspace(-4, 4, 100)
y = x**2

# Pyplot approach
plt.plot(x, y)  # Plot the function

# Add title and axis labels
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Create a figure with 2 subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Scatter plot on the first subplot
axs[0].scatter(random_data_x, random_data_y, color='blue', alpha=0.5)
axs[0].set_title('Scatter Plot')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Bar plot on the second subplot
axs[1].bar(names, values, color='green')
axs[1].set_title('Bar Plot')
axs[1].set_xlabel('Fruit')
axs[1].set_ylabel('Quantity')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
