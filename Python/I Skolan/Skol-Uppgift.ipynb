```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

skol_data = pd.read_excel(r'skoStorlek_screentime.xlsx')

skol_data.head()

# Rensa data för att undvika problem med NaN
shoe_sizes = skol_data['Skostorlek'].dropna()
screentimes = skol_data['Skärmanvändning'].dropna()

# Medelvärde för Skostorlek
mean_shoe_size = shoe_sizes.mean()  
print(f"Medelvärde för Skostorlek: {mean_shoe_size}")

# --- Graf 1: Skostorlek ---
plt.figure(figsize=(8, 4))
plt.bar(shoe_sizes.index, shoe_sizes, color='skyblue')
plt.axhline(mean_shoe_size, color='red', linestyle='--', label=f'Medelvärde: {mean_shoe_size:.2f}')
plt.title('Skostorlek')
plt.xlabel('Index')
plt.ylabel('Skostorlek')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Graf 2: Skärmanvändning ---

# Räkna ut medelvärde
average_time = screentimes.mean()
print(f"Genomsnittlig skärmtid: {average_time:.2f} timmar")

# Plotta
plt.figure(figsize=(10, 5))
plt.plot(screentimes.index, screentimes, marker='o', color='green', label='Skärmtid')
plt.axhline(y=average_time, color='red', linestyle='--', label=f'Medel: {average_time:.2f} h')

# Layout och etiketter
plt.title('Skärmanvändning över tid')
plt.xlabel('Datum')
plt.ylabel('Timmar')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)  # Roterar datumen för bättre läsbarhet
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.boxplot(screentimes, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.axhline(y=average_time, color='red', linestyle='--', label=f'Medel: {average_time:.2f} h')

plt.title("Boxplot över skärmtid")
plt.ylabel("Timmar")
plt.legend()
plt.grid(True)
plt.show()
```
